## Conviertiendo las fechas de tipo Object a datetime

```python
fechas = ['fecha_fallecimiento', 'fecha_dosis1', 'fecha_dosis2', 'fecha_dosis3', 'fecha_ingreso_hosp', 'fecha_ingreso_uci', 'fecha_ingreso_ucin', 'fecha_segumiento_hosp_ultimo']

for fecha in fechas:
    # print(fecha)
    covid[fecha] = pd.to_datetime(covid[fecha], dayfirst=True)
```

```python
departamentos = covid.groupby(['dpt_cdc'])[['edad']].describe()
```

```python
departamentos.columns = departamentos.columns.droplevel(0)
```

```python
departamentos
```

```python
f, ax = plt.subplots(figsize=(16,6))

ax.bar(
    x = departamentos.index,
    height = departamentos['mean'],
    color = 'green'
)

ax.set(title='Media de edad por departamento')

plt.xticks(rotation = 45, ha = 'right')
plt.show()
plt.close()
```
